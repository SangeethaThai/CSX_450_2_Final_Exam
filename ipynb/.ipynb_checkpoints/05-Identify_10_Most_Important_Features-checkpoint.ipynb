## Identify 10 Most Important Features

```r
source('../src/util.r')
source('../src/multiplot.r')
```

```r
ames_housing_df <- loadCleanAmesData()
```

```r
features_df <- ames_housing_df[, !names(ames_housing_df) %in% c("SalePrice")]
```

```r
dim(features_df)
```

```r
head(features_df)
```

#### Let's count the number of 0s in the columns. 

```r
zero_sums <- colSums(features_df == 0)
zero_sums <- zero_sums[zero_sums > 0]
#zero_sums[zero_sums < 1460 * 0.50]
zero_sums
```

#### Zero's in these columns indicate that the feature is not available in those homes. If a feature is not available in more than 50% of the homes, we can assume the features are not likely to affect the SalePrice prediction and hence remove those features from consideration.

```r
features_df <- features_df[, colSums(features_df == 0) < (1460 * 0.50)]
```

```r
dim(features_df)
```

#### If PoolArea is irrelevant as above, Pool Quality is irrelevant for our prediction and can be removed as well.

```r
features_df <- features_df[, !names(features_df) %in% c("PoolQC")]
```

```r
dim(features_df)
```

#### Split numerical and categorical features

```r
numerical_features <- Filter(is.numeric, features_df)
categorical_features <- Filter(is.factor, features_df)
```

```r
dim(numerical_features)
dim(categorical_features)
```

#### Log and scale the numerical features

```r
numerical_features = log(numerical_features + 1)
```

```r
numerical_features = data.frame(scale(numerical_features))
```

#### One Hot Encode Categorical features

```r
install.packages("onehot")
```

```r
library("onehot")
```

```r
one_hot_encoding = onehot(categorical_features, stringsAsFactors = FALSE, addNA = FALSE, max_levels = 25) 
encoded_categorical_features <- predict(one_hot_encoding, categorical_features)
```

```r
dim(encoded_categorical_features)
```

```r
head(encoded_categorical_features)
```

### Combining Correlation with the significance test (p > 0.01) to choose the 10 most important features 

#### (a) Choose from Numeric Features first

```r
numerical_features_with_target <- numerical_features
numerical_features_with_target$SalePrice <- ames_housing_df$SalePrice
numerical_features_with_target$Id <- NULL
cor_features = cor(numerical_features_with_target)
cor_features
```

```r
install.packages("corrplot")
```

```r
library("corrplot")
```

```r
cor.mtest <- function(mat) {
    mat <- as.matrix(mat)
    n <- ncol(mat)
    p.mat<- matrix(NA, n, n)
    diag(p.mat) <- 0
    for (i in 1:(n - 1)) {
        for (j in (i + 1):n) {
            tmp <- cor.test(mat[, i], mat[, j])
            p.mat[i, j] <- p.mat[j, i] <- tmp$p.value
        }
    }
  colnames(p.mat) <- rownames(p.mat) <- colnames(mat)
  p.mat
}
# matrix of the p-value of the correlation
p.mat <- cor.mtest(numerical_features_with_target)
```

```r
# Specialized the insignificant value according to the significant level
corrplot(cor_features, type="upper", order="hclust", 
         p.mat = p.mat, sig.level = 0.01)
```

**From the plot, let's choose 5 most significant features**
- GrLivArea
- FirstFlrSF
- YearBuilt
- YearRemodAdd
- GarageYeBuilt

#### Choose from one-hot encoded categorical features

```r
categorical_sums = colSums(encoded_categorical_features)
categorical_sums <- categorical_sums[categorical_sums > 0]
sort(categorical_sums, decreasing=TRUE)
```

#### Let's select oe hot encoded features with the most frequency, arguing that higher the number, greater the significance and impact. Based on this theory, the most significant categorical features are
- Utilities
- Street
- Condition2 (Proximity to various conditions)
- RoofMatl
- Heating
